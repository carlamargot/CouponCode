### Merging Train data with campagin data 

```python
#pulling in the campaign data 
campaign_df = pd.read_csv('campaign_data.csv')
campaign_df.drop(columns=['start_date','end_date'], inplace=True)
campaign_df
```

```python
train_df = pd.read_csv('train.csv')
df = pd.merge(train_df,campaign_df,left_on='campaign_id',right_on='campaign_id',how='outer')
df.set_index('campaign_id')
df.dropna(subset=['customer_id'],inplace=True)
df
```

```python
# merging with new_demo 
practice = pd.merge(new_demo,df,left_on='customer_id',right_on='customer_id',how='inner')
```

```python
# looking at why the data is so big 
practice.category.value_counts()
```

```python
#dropping the indexes that do not belong to prepared food so we only have to work with prepared food data
practice.drop(practice[practice['category']!='Prepared Food'].index,inplace=True)
```

```python
practice.drop(columns=['campaign_id'],axis=1,inplace=True)
```

```python
practice.drop(columns=['id'],axis=1,inplace=True)
```

```python
practice.set_index('customer_id')
```

```python
# checking to see how many duplicates exist
practice.duplicated().value_counts()
```

```python
practice.drop_duplicates(inplace=True)
```
